## Display the details of top 5 student based on their marks

```python
import pandas as pd
import numpy as np
# Function to generate random student names
def generate_random_student_name():
    first_names = ['John', 'Jane', 'Michael', 'Emily', 'David', 'Emma', 'Chris', 'Olivia', 'Sophia', 'Daniel']
    last_names = ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Taylor', 'Clark']
    return np.random.choice(first_names) + ' ' + np.random.choice(last_names)
# Create a list of random student names
random_students = [generate_random_student_name() for _ in range(10)]
# Create a dictionary with student details and random subject scores
data = {
    'Student': random_students,
    'TAMIL': np.random.randint(25, 101, size=10),
    'ENGLISH': np.random.randint(25, 101, size=10),
    'MATHS': np.random.randint(25, 101, size=10),
    'SCIENCE': np.random.randint(25, 101, size=10),
    'CS': np.random.randint(25, 101, size=10)
}
# Create the DataFrame
df = pd.DataFrame(data)
# Set the 'Student' column as the index
df.set_index('Student', inplace=True)
# Add a column for Total Marks of each student
df['Total Marks'] = df[['TAMIL', 'ENGLISH', 'MATHS', 'SCIENCE', 'CS']].sum(axis=1)
# Sort the DataFrame by 'Total Marks' in descending order
df_sorted = df.sort_values(by='Total Marks', ascending=False)
# Display the details of the top 5 students based on marks
top_5_students = df_sorted.head(5)
print(top_5_students)
```
